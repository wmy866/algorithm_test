Selection sort (選擇排序)

```
def Selction_Sort(datas):
    for i in range(len(datas)-1):                                 # n-1 次
        min = i # 第一次選從第一個數開始,第二次從第二個數開始                # n-1 次
        for j in range(i+1, len(datas)):    # 用記錄的max數與其他的比較         # (n-1)^2 次
            if datas[j] < datas[min]:                           # (n-1)^2 次
                min = j                                # (n-1)^2 次

        datas[i], datas[min] = datas[min], datas[i]                     # n-1 次

        print(datas)                                       # n-1 次
        print('---'*20)                                     # n-1 次
    return datas                                          # 1 次
                                                  #O(n^2) 

if __name__ == '__main__':
    import random
    a = list(range(16))
    random.shuffle(a)
    print(a)
    a = Selction_Sort(a)
    print(a)
```
